### Duas soluções para o problema

#### Solução 01

```python
cota_inicial = 0
pap = 3
n_pav = 30



lista_pav = []
for i in range(n_pav + 1):
  cota_atual =   cota_inicial + (pap * i)
  cota_atual = round(cota_atual, 2)
  # criar caixa na cota atual
  box = cq.Workplane("XY").box(30 , 40, pap, centered=[True, True, False]).translate((0, 0, cota_atual))
  lista_pav.append(box)
  # print(f"Cota do pavimento {i} = {cota_atual}")
print(lista_pav)

show(lista_pav)
```

#### Solução 02

```python
import cadquery as cq

cota_inicial = 0
pap = 3
n_pav = 30

lista_pav = []
for i in range(n_pav + 1):
    cota_atual = round(cota_inicial + (pap * i), 2)

    pav = (
        cq.Workplane(
                    "XY", 
                     origin=(0,0, cota_atual) # desloca o plano de desenho para a cota atula, não precisa usar translate
                     ) 
        .rect(30, 40)        # desenha o retângulo no plano XY
        .extrude(pap)        # extrusão em Z pelo valor de pap
    )

    lista_pav.append(pav)
    # print(f"Cota do pavimento {i} = {cota_atual}")

show(lista_pav)
```

#### Variação com scale

```python
import numpy as np

cota_inicial = 0
pap = 3
n_pav = 30

smooth_factor = 1/(np.pi * 2)

lista_pav = []
for i in range(n_pav + 1):
    cota_atual = round(cota_inicial + (pap * i), 2)
    scale_factor = np.abs(np.sin(i) * smooth_factor + 1)

    pav = (
        cq.Workplane("XY", origin=(0, 0, cota_atual))  # plano de trabalho já na cota correta
        .rect(30 * scale_factor, 40 * scale_factor)
        .extrude(pap)
        
    )

    lista_pav.append(pav)
    # print(f"Cota do pavimento {i} = {cota_atual:.2f} | scale = {scale_factor:.4f}")

show(lista_pav)
```

### Rotação incremental


#### Solução 01

```python
# parâmetros
cota_inicial = 0
pap = 3
n_pav = 30
rot_inc = 2.5

# code
lista_pav = []
for i in range(n_pav + 1):
  cota_atual =   cota_inicial + (pap * i)
  cota_atual = round(cota_atual, 2)

  box = cq.Workplane("XY").box(30 , 40, pap, centered=[True, True, False]).translate((0, 0, cota_atual))

  box = box.rotate((0,0,0), (0,0,1), i * rot_inc)
  lista_pav.append(box)
  # print(f"Cota do pavimento {i} = {cota_atual}")

# show
show(lista_pav)
```

#### Solução 02

```python
# parâmetros
cota_inicial = 0
pap = 3
n_pav = 30
rot_inc = 2.5

# code
lista_pav = []
for i in range(n_pav + 1):
    cota_atual = round(cota_inicial + (pap * i), 2)
    scale_factor = np.abs(np.sin(i) * smooth_factor + 1)

    pav = (
        cq.Workplane(origin=(0, 0, cota_atual))
        .rect(30, 40)
        .extrude(pap)
        .rotate((0, 0, 0), (0, 0, 1), i * rot_inc)
    )

    lista_pav.append(pav)

# show
show(lista_pav)
```

#### Solução 02 com fator de escalonamento

```python
# parâmetros
cota_inicial = 0
pap = 3
n_pav = 30
rot_inc = 2.5

smooth_factor = 1/(np.pi * 2)


# code
lista_pav = []
for i in range(n_pav + 1):
    cota_atual = round(cota_inicial + (pap * i), 2)
    scale_factor = np.abs(np.sin(i) * smooth_factor + 1)

    pav = (
        cq.Workplane(origin=(0, 0, cota_atual))
        .rect(30 * scale_factor, 40 * scale_factor)
        .extrude(pap)
        .rotate((0, 0, 0), (0, 0, 1), i * rot_inc)
    )

    lista_pav.append(pav)

# show
show(lista_pav)
```

## Exportando o modelo

```python
# Create an assembly and add the objects
assy = cq.Assembly()
for obj in lista_pav:
    assy.add(obj)

# Export the assembly to a STEP file
assy.export(
    "output.step", 
    mode="default" # ou mode = "fused" para extrair os objetos unidos (solid union)
    )
```
