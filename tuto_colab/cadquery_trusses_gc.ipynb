## Revisão: Listas

Uma **lista** é uma coleção ordenada de valores. Em Python, listas são criadas com colchetes `[]` e podem conter qualquer tipo de dado — números, textos, outros objetos.

### Criação de listas

```python
# Lista de números
cotas = [0.0, 3.0, 6.0, 9.0, 12.0]
print(cotas)

# Lista de textos
nomes = ["nó A", "nó B", "nó C"]
print(nomes)

# Lista vazia — será preenchida depois
barras = []
print(barras)
```

### Índices — acessando elementos

Cada elemento de uma lista tem um **índice** numérico que começa em `0`. Índices negativos contam do final da lista.

```python
cotas = [0.0, 3.0, 6.0, 9.0, 12.0]

print(cotas[0])   # primeiro elemento → 0.0
print(cotas[2])   # terceiro elemento  → 6.0
print(cotas[-1])  # último elemento    → 12.0
print(cotas[-2])  # penúltimo          → 9.0

# Modificando um elemento pelo índice
cotas[1] = 3.5
print(cotas)
```

### `.append()` — adicionando ao final

```python
barras = []

barras.append("barra 0")
barras.append("barra 1")
barras.append("barra 2")

print(barras)
print(f"Total de barras: {len(barras)}")
```

### `.insert()` — adicionando em uma posição específica

```python
cotas = [0.0, 6.0, 12.0]
print("Antes:", cotas)

# insert(índice, valor) → insere antes do elemento naquele índice
cotas.insert(1, 3.0)   # insere 3.0 na posição 1
cotas.insert(3, 9.0)   # insere 9.0 na posição 3

print("Depois:", cotas)
```

### `.remove()` — removendo pelo valor

```python
nomes = ["nó A", "nó B", "nó X", "nó C"]
print("Antes:", nomes)

# remove a primeira ocorrência do valor informado
nomes.remove("nó X")

print("Depois:", nomes)
```

### `del` — removendo pelo índice

```python
cotas = [0.0, 3.0, 99.0, 6.0, 9.0]
print("Antes:", cotas)

# del remove o elemento no índice informado
del cotas[2]

print("Depois:", cotas)
```

### Soma: números, textos e listas

O operador `+` tem comportamentos diferentes dependendo do tipo:

```python
# Soma de números → resultado numérico
a = 3
b = 4
print(a + b)          # 7

# Soma de textos (strings) → concatenação
s1 = "nó "
s2 = "superior"
print(s1 + s2)        # "nó superior"

# Soma de listas → concatenação de listas
lista_a = [1, 2, 3]
lista_b = [4, 5, 6]
print(lista_a + lista_b)  # [1, 2, 3, 4, 5, 6]

# ATENÇÃO: não dá para somar número com texto
# print(a + s1)  # → TypeError!
```

## Revisão: Comparações Lógicas

Comparações retornam `True` (verdadeiro) ou `False` (falso) e são a base de qualquer decisão no código.

| Operador | Significado |
|----------|-------------|
| `==` | igual a |
| `!=` | diferente de |
| `>`  | maior que |
| `<`  | menor que |
| `>=` | maior ou igual a |
| `<=` | menor ou igual a |
| `%`  | resto da divisão (módulo) |
| `//` | quociente inteiro (divisão inteira) |

```python
a = 10
b = 3

print(a == 10)   # True  — a é igual a 10?
print(a != b)    # True  — a é diferente de b?
print(a >= b)    # True  — a é maior ou igual a b?
print(a <= b)    # False — a é menor ou igual a b?

print(a % b)     # 1     — resto de 10 ÷ 3
print(a // b)    # 3     — quociente inteiro de 10 ÷ 3
```

### Par, ímpar e divisibilidade

O **módulo** (`%`) é a ferramenta fundamental para testar divisibilidade:
- `n % 2 == 0` → `n` é par
- `n % 2 != 0` → `n` é ímpar

```python
n = 12   # experimente trocar por outros valores

# Teste par ou ímpar
if n % 2 == 0:
    print(f"{n} é par")
else:
    print(f"{n} é ímpar")
```

```python
n = 12

# Teste par E divisível por 4
if n % 2 == 0 and n % 4 == 0:
    print(f"{n} é par e divisível por 4")
elif n % 2 == 0:
    print(f"{n} é par, mas NÃO é divisível por 4")
else:
    print(f"{n} é ímpar")
```

```python
n = 15

# Teste ímpar E divisível por 5
if n % 2 != 0 and n % 5 == 0:
    print(f"{n} é ímpar e divisível por 5")
elif n % 2 != 0:
    print(f"{n} é ímpar, mas NÃO é divisível por 5")
else:
    print(f"{n} é par")
```

## Revisão: Funções

Uma **função** é um bloco de código nomeado que pode ser reutilizado. Em Python, funções são definidas com a palavra-chave `def`.

Existe uma diferença importante entre funções que **imprimem** um resultado e funções que **retornam** um resultado.

### Função que imprime — sem `return`

A função executa uma ação (imprimir), mas não produz um valor que pode ser armazenado ou usado em outro cálculo.

```python
def soma_print(a, b):
    print(a + b)

soma_print(3, 4)     # imprime 7

# Tentar guardar o resultado não funciona:
resultado = soma_print(3, 4)
print(resultado)     # None — a função não retornou nada
```

### Função que retorna — com `return`

A função produz um valor que pode ser armazenado em uma variável, usado em outro cálculo, ou passado para outra função.

```python
def soma(a, b):
    return a + b

resultado = soma(3, 4)
print(resultado)          # 7

# O valor retornado pode ser usado diretamente em outros cálculos
print(soma(3, 4) * 2)     # 14
print(soma(soma(1, 2), 3)) # 6
```

### Resumo da diferença

| | `print()` | `return` |
|-|-----------|----------|
| Exibe na tela | ✅ | ❌ (a menos que você imprima depois) |
| Pode ser guardado em variável | ❌ | ✅ |
| Pode ser usado em outro cálculo | ❌ | ✅ |
| Valor fora da função | `None` | o valor retornado |

> 💡 **Regra prática**: funções que **produzem dados** para serem usados no programa devem ter `return`. Funções que apenas **reportam** algo ao usuário podem usar `print`.

## Pontos no Espaço 3D

No CadQuery, pontos no espaço são representados por objetos `cq.Vector`. Cada vetor armazena três coordenadas: `x`, `y` e `z`.

A função `show()` da cadquery-simpleViewer exibe `cq.Vector` diretamente como marcadores no visualizador 3D — sem necessidade de criar sólidos.

Vamos criar quatro pontos que esboçam o perfil de uma treliça de cobertura:

```python
# Quatro pontos definindo um perfil simples
p1 = cq.Vector( 0, 0, 0)
p2 = cq.Vector( 3, 0, 2)
p3 = cq.Vector( 7, 0, 2)
p4 = cq.Vector(10, 0, 0)

show(
    [p1, p2, p3, p4],
    names=["P1", "P2", "P3", "P4"],
    points_display=dict(size=10, color="steelblue", symbol="circle")
)
```

## Polilinha

Uma **polilinha** é uma sequência de segmentos de linha conectando pontos em ordem. Em CadQuery, um conjunto de segmentos conectados é representado por um `cq.Wire`.

O método `cq.Wire.makePolygon()` recebe uma lista de `cq.Vector` e cria um wire com segmentos retos entre eles. A cadquery-simpleViewer exibe `cq.Wire` diretamente como linha, com o parâmetro `lines_display` controlando a aparência.

```python
# Criando a polilinha como um wire CadQuery
polilinha = cq.Wire.makePolygon([p1, p2, p3, p4])

# Exibindo a polilinha e os nós na mesma cena
show(
    [polilinha, p1, p2, p3, p4],
    names=["Polilinha", "P1", "P2", "P3", "P4"],
    lines_display=dict(color="steelblue", width=4),
    points_display=dict(size=8, color="indianred", symbol="circle")
)
```

## Treliça — Definição dos Banzos

A treliça plana é composta por dois elementos lineares principais chamados **banzos**:

- **Banzo superior** (`banzo_sup`): o elemento de cima, que pode ser reto ou arqueado
- **Banzo inferior** (`banzo_inf`): o elemento de baixo, paralelo ao superior

No CadQuery, usamos `cq.Edge` para criar elementos lineares (retos ou curvos):

| Método | Descrição |
|--------|-----------|
| `cq.Edge.makeLine(v1, v2)` | Segmento de linha reto entre dois pontos |
| `cq.Edge.makeThreePointArc(v1, v2, v3)` | Arco circular passando por três pontos |

### Parâmetros da treliça

```python
# Parâmetros gerais da treliça
comprimento  = 10.0   # comprimento total em metros
z_sup        = 12.0   # cota do banzo superior (m)
altura_treli = 1.0    # distância vertical entre banzos (m)
n_paineis    = 6      # número de painéis (divisões) da treliça
```

### Banzo Superior

O banzo superior pode ser definido como uma **linha reta** ou como um **arco** no plano XZ. As duas opções estão no código abaixo — comente uma e descomente a outra para alternar entre os dois perfis:

- **Linha reta**: adequada para treliças com cobertura plana ou levemente inclinada
- **Arco**: adequada para treliças com cobertura curva (treliça Pratt, Howe, etc.)

```python
# --- Opção 1: Banzo superior reto ---
banzo_sup = cq.Edge.makeLine(
    cq.Vector(0,          0, z_sup),
    cq.Vector(comprimento, 0, z_sup)
)

# --- Opção 2: Banzo superior em arco ---
# Descomente as linhas abaixo e comente a Opção 1 para usar o arco.
# O ponto do meio define a flecha (altura) do arco.
# flecha = 1.5
# banzo_sup = cq.Edge.makeThreePointArc(
#     cq.Vector(0,               0, z_sup),
#     cq.Vector(comprimento / 2, 0, z_sup + flecha),
#     cq.Vector(comprimento,     0, z_sup)
# )
```

### Banzo Inferior

O banzo inferior é uma cópia do banzo superior deslocada `altura_treli` metros para baixo no eixo Z.

```python
# Cópia do banzo superior deslocada para baixo
banzo_inf = copy(banzo_sup)
banzo_inf = banzo_inf.translate(cq.Vector(0, 0, -altura_treli))
```

---

## Geração dos Nós

Os **nós** da treliça são os pontos de conexão entre as barras. Eles são gerados dividindo os banzos em `n_paineis` espaços iguais usando o método `.positionAt(t)`, que retorna um `cq.Vector` para qualquer parâmetro `t` entre `0.0` e `1.0` ao longo do elemento.

Isso funciona tanto para elementos retos quanto para arcos — o método sempre percorre a geometria real do elemento.

```python
def gerar_nos(aresta, n):
    """
    Divide uma aresta em n espaços iguais e retorna uma lista de cq.Vector.

    aresta : cq.Edge — aresta a dividir
    n      : int     — número de painéis (divisões)
    """
    nos = []
    for i in range(n + 1):
        t = i / n
        nos.append(aresta.positionAt(t))
    return nos


# Nós do banzo superior e inferior como listas de cq.Vector
points_sup = gerar_nos(banzo_sup, n_paineis)
points_inf = gerar_nos(banzo_inf, n_paineis)

print(f"Nós gerados no banzo superior: {len(points_sup)}")
print(f"Nós gerados no banzo inferior: {len(points_inf)}")

print("\nBanzo superior — coordenadas dos nós:")
for i in range(len(points_sup)):
    p = points_sup[i]
    print(f"  [{i}]  x={p.x:.3f}  y={p.y:.3f}  z={p.z:.3f}")
```

## Visualização dos Banzos e Nós

Passamos os dois banzos (`cq.Edge`) e as duas listas de nós (`cq.Vector`) diretamente para `show()`. O parâmetro `lines_display` controla a aparência das arestas e `points_display` controla os marcadores.

> ⚠️ **Nota**: os estilos se aplicam uniformemente a todos os elementos do mesmo tipo na chamada. Para estilos diferentes por grupo, use chamadas `show()` separadas.

```python
# Construindo a lista de objetos e nomes
objetos = [banzo_sup, banzo_inf]
nomes   = ["Banzo superior", "Banzo inferior"]

for i in range(len(points_sup)):
    objetos.append(points_sup[i])
    nomes.append("NS" + str(i))

for i in range(len(points_inf)):
    objetos.append(points_inf[i])
    nomes.append("NI" + str(i))

show(
    objetos,
    names=nomes,
    lines_display=dict(color="steelblue", width=4, samples=60),
    points_display=dict(size=8, color="indianred", symbol="circle")
)
```

> 💡 **Observe**: os banzos são exibidos como linhas e os nós como marcadores — tudo com uma única chamada `show()`, sem nenhum código Plotly manual.
