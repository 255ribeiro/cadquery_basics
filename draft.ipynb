# Draft - Plotly Sliders Examples

Este notebook reúne os exemplos gerados na conversa para testes com sliders.

```python
# Exemplo 1: ipywidgets (Int + Float) controlando FigureWidget
import numpy as np
import plotly.graph_objects as go
import ipywidgets as widgets
from IPython.display import display

x = np.linspace(0, 2 * np.pi, 500)

fig = go.FigureWidget(
    data=[go.Scatter(x=x, y=np.sin(x), mode="lines", name="y = A * sin(f*x)")],
    layout=go.Layout(
        title="Exemplo Plotly com slider inteiro + float",
        template="plotly_white",
        xaxis_title="x",
        yaxis_title="y",
        height=420
    )
)

freq_slider = widgets.IntSlider(
    value=1, min=1, max=12, step=1, description="freq (int)"
)

amp_slider = widgets.FloatSlider(
    value=1.0, min=0.1, max=3.0, step=0.1, description="amp (float)"
)

def atualizar(freq, amp):
    y = amp * np.sin(freq * x)
    with fig.batch_update():
        fig.data[0].y = y
        fig.layout.title = f"freq={freq} | amp={amp:.1f}"

widgets.interactive_output(
    atualizar,
    {"freq": freq_slider, "amp": amp_slider}
)

display(widgets.VBox([freq_slider, amp_slider]), fig)
```

```python
# Exemplo 2: sliders nativos do Plotly (layout.sliders)
import numpy as np
import plotly.graph_objects as go

x = np.linspace(0, 2 * np.pi, 600)
freqs = list(range(1, 9))
amps = [0.5, 0.8, 1.0, 1.2, 1.5, 2.0, 2.5]
amp0 = 1.0

def make_y(a, f):
    return (a * np.sin(f * x)).tolist()

fig = go.Figure()

for i, f in enumerate(freqs):
    fig.add_trace(
        go.Scatter(
            x=x,
            y=make_y(amp0, f),
            mode="lines",
            name=f"freq={f}",
            visible=(i == 0)
        )
    )

freq_steps = []
for i, f in enumerate(freqs):
    vis = [False] * len(freqs)
    vis[i] = True
    step = dict(
        method="update",
        args=[
            {"visible": vis},
            {"title": f"Senoide | freq={f} | amp (slider de baixo)"}
        ],
        label=str(f)
    )
    freq_steps.append(step)

freq_slider = dict(
    active=0,
    currentvalue={"prefix": "freq (int): "},
    pad={"t": 30},
    x=0.1, y=0.12, len=0.8,
    steps=freq_steps
)

amp_steps = []
for a in amps:
    y_all = [make_y(a, f) for f in freqs]
    step = dict(
        method="restyle",
        args=[{"y": y_all}],
        label=f"{a:.1f}"
    )
    amp_steps.append(step)

amp_slider = dict(
    active=amps.index(amp0),
    currentvalue={"prefix": "amp (float): "},
    pad={"t": 90},
    x=0.1, y=0.0, len=0.8,
    steps=amp_steps
)

fig.update_layout(
    template="plotly_white",
    title="Senoide | freq=1 | amp (slider de baixo)",
    xaxis_title="x",
    yaxis_title="y",
    height=520,
    sliders=[freq_slider, amp_slider]
)

fig.show()
```
